Frequency distribution of users by device type

```python
# histogram of users of device type
plt.figure(1)
signups['device'].hist(bins=7,color = 'red')
plt.title('Frequency distribution of users by device type',y=1.02)
plt.xlabel('device type')
plt.ylabel('number of users')
plt.grid()

savefig('freqdistr_device_type.pdf', dpi=None, facecolor='w', edgecolor='w',
        orientation='portrait', papertype='letter', format='pdf',
        transparent=False, bbox_inches='tight', pad_inches=0.1,
        frameon=None)
plt.show()
```

```python
#replacing the auth_type A,B,C with 1,2,3 for integer data
signups.loc[signups['auth_type']=='A','auth_type']=1.0
signups.loc[signups['auth_type']=='B','auth_type']=2.0
signups.loc[signups['auth_type']=='C','auth_type']=3.0
```

```python
signups.head(5)
```

```python
signups['auth_type'] = signups['auth_type'].astype('int64')
```

```python
n_auth_type = np.array(signups['auth_type'].value_counts())
```

```python
signups['auth_type'].value_counts()
```

Frequency distribution by auth type

```python
y = [n_auth_type[1],n_auth_type[2],n_auth_type[0]]
labels = ['A','B','C']
ngroups = 3
index = np.arange(ngroups)
barwidth = 0.3

plt.figure(2)
plt.bar(index, y, barwidth, color = 'red')
plt.xticks(index + 0.15, labels)
plt.title('Frequency distribution of users by auth type')
plt.ylabel('number of users')
plt.xlabel('auth type')
plt.margins(0.1)
plt.show()
```

Frequency distribution of users by combination of device and auth type

```python
signups_device = signups.groupby('device')
labels = ['A','B','C']
ngroups = 3
index = np.arange(ngroups)
barwidth = 0.3

plt.figure(3,figsize=(18,10))

plt.subplot(2,3,1)
temp = signups_device.get_group(1)
temp_counts = np.array(temp['auth_type'].value_counts().sort_values())
y = [temp_counts[0],temp_counts[1],temp_counts[2]]
plt.bar(index, y, barwidth, color = 'red',label = 'device type:1')
plt.xticks(index + 0.15, labels)
plt.legend(loc = 2)
plt.ylabel('number of users')
plt.xlabel('auth_type')
plt.margins(0.1)

plt.subplot(2,3,2)
temp = signups_device.get_group(2)
temp_counts = np.array(temp['auth_type'].value_counts().sort_values())
y = [temp_counts[0],temp_counts[1],temp_counts[2]]
plt.bar(index, y, barwidth, color = 'red',label = 'device type:2')
plt.xticks(index + 0.15, labels)
plt.legend(loc = 2)
plt.ylabel('number of users')
plt.xlabel('auth_type')
plt.margins(0.1)

plt.subplot(2,3,3)
temp = signups_device.get_group(3)
temp_counts = np.array(temp['auth_type'].value_counts().sort_values())
y = [temp_counts[0],temp_counts[1],temp_counts[2]]
plt.bar(index, y, barwidth, color = 'red',label = 'device type:3')
plt.xticks(index + 0.15, labels)
plt.legend(loc = 2)
plt.ylabel('number of users')
plt.xlabel('auth_type')
plt.margins(0.1)

plt.subplot(2,3,4)
temp = signups_device.get_group(4)
temp_counts = np.array(temp['auth_type'].value_counts().sort_values())
y = [temp_counts[0],temp_counts[1],temp_counts[2]]
plt.bar(index, y, barwidth, color = 'red',label = 'device type:4')
plt.xticks(index + 0.15, labels)
plt.legend(loc = 2)
plt.ylabel('number of users')
plt.xlabel('auth_type')
plt.margins(0.1)

plt.subplot(2,3,5)
temp = signups_device.get_group(5)
temp_counts = np.array(temp['auth_type'].value_counts().sort_values())
y = [temp_counts[0],temp_counts[1],temp_counts[2]]
plt.bar(index, y, barwidth, color = 'red',label = 'device type:5')
plt.xticks(index + 0.15, labels)
plt.legend(loc = 2)
plt.ylabel('number of users')
plt.xlabel('auth_type')
plt.margins(0.1)

plt.subplot(2,3,6)
temp = signups_device.get_group(6)
temp_counts = np.array(temp['auth_type'].value_counts().sort_values())
y = [temp_counts[0],temp_counts[1],temp_counts[2]]
plt.bar(index, y, barwidth, color = 'red',label = 'device type:6')
plt.xticks(index + 0.15, labels)
plt.legend(loc = 2)
plt.ylabel('number of users')
plt.xlabel('auth_type')
plt.margins(0.1)
plt.suptitle('Number of users for different auth_type segmented by device type',fontsize = 20)

savefig('bar_device_type.pdf', dpi=None, facecolor='w', edgecolor='w',
        orientation='portrait', papertype='letter', format='pdf',
        transparent=False, bbox_inches='tight', pad_inches=0.1,
        frameon=None)
plt.show()
```
